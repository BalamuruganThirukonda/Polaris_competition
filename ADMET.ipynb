Calculate Lipinski descriptors

```python
import numpy as np
from rdkit import Chem
from rdkit.Chem import Descriptors, Lipinski
```

Calculate descriptors

```python
def lipinski(smiles, verbose=False):
  moldata = []
  for elem in smiles:
    mol = Chem.MolFromSmiles(elem)
    moldata.append(mol)

  baseData = np.arange(1, 1)
  i = 0
  for mol in moldata:
    desc_MolWt = Descriptors.MolWt(mol)
    desc_MolLogP = Descriptors.MolLogP(mol)
    desc_NumHDonors = Lipinski.NumHDonors(mol)
    desc_NumHAcceptors = Lipinski.NumHAcceptors(mol)

    row = np.array([desc_MolWt,
                    desc_MolLogP,
                    desc_NumHDonors,
                    desc_NumHAcceptors])
    if i == 0:
      baseData = row
    else:
      baseData = np.vstack([baseData, row])
    i = i + 1

  columnNames = ["MW", "LogP", "NumHDonors", "NumHAcceptors"]
  descriptors = pd.DataFrame(data=baseData, columns=columnNames)

  return descriptors
```

LogP follows normal distribution, other columns having skewness in their data. So, instead of filling missing value with mean or median, we analyse the structure of the drug and fill the value of missing drug properties based on the highest similarity match with other drugs structure.

```python
from rdkit import Chem
from rdkit.Chem import AllChem
from rdkit.DataStructs import TanimotoSimilarity
import pandas as pd
import numpy as np

# Helper function to compute Tanimoto similarity
def tanimoto_similarity(fp1, fp2):
    if fp1 is None or fp2 is None:
        return 0
    return TanimotoSimilarity(fp1, fp2)

# Function to calculate fingerprints for each molecule
def generate_fingerprints(df):
    fingerprints = []
    for smiles in df["smiles"]:
        mol = Chem.MolFromSmiles(smiles)
        if mol:
            fingerprints.append(AllChem.GetMorganFingerprintAsBitVect(mol, 2, nBits=2048))  # Morgan Fingerprint
        else:
            fingerprints.append(None)
    df["fingerprint"] = fingerprints
    return df

# Function to fill missing values based on similarity
def fill_missing_values(df, properties):
    for prop in properties:
        missing_rows = df[df[prop].isna()].index  # Find rows with missing values

        for missing_idx in missing_rows:
            missing_fp = df.loc[missing_idx, "fingerprint"]

            # Compute similarities with non-missing rows
            similarities = []
            for i, row in df.dropna(subset=[prop]).iterrows():
                sim = tanimoto_similarity(missing_fp, row["fingerprint"])
                if sim > 0.5:  # Only consider similarities greater than 90%
                    similarities.append((i, sim, row[prop]))

            # Sort by similarity (highest first)
            similarities.sort(key=lambda x: x[1], reverse=True)

            # Check if there are valid matches with similarity > 80%
            if similarities:
                best_match_value = similarities[0][2]  # Get the most similar property's value
                df.at[missing_idx, prop] = best_match_value
            else:
                print(f"Skipping: No valid match (sim > 80%) for {prop} at index {missing_idx}")

    return df  # Return the modified DataFrame
```

```python
# Apply to all property columns
train_df = generate_fingerprints(train_df)
property_columns = ["HLM", "MDR1-MDCKII", "MLM", "LogD", "KSOL"]  # List of property columns
fill_missing_values(train_df, property_columns)
```

```python
def euclidean_similarity(desc1, desc2):
    return np.linalg.norm(np.array(desc1) - np.array(desc2))

# Function to fill missing values based on Lipinski descriptors
def fill_missing_values_lipinski(df, properties):
    for prop in properties:
        missing_rows = df[df[prop].isna()].index  # Find rows with missing values

        for missing_idx in missing_rows:
            missing_desc = df.loc[missing_idx, ["MW", "LogP", "NumHDonors", "NumHAcceptors"]].values

            # Compute similarities with non-missing rows based on Lipinski descriptors
            similarities = []
            for i, row in df.dropna(subset=[prop]).iterrows():
                row_desc = row[["MW", "LogP", "NumHDonors", "NumHAcceptors"]].values
                sim = euclidean_similarity(missing_desc, row_desc)
                similarities.append((i, sim, row[prop]))

            # Sort by similarity (lowest distance first, as we use Euclidean)
            similarities.sort(key=lambda x: x[1])

            # Check if there are valid matches (similarity is not infinity)
            if similarities:
                best_match_value = similarities[0][2]  # Get the most similar property's value
                df.at[missing_idx, prop] = best_match_value
            else:
                print(f"Skipping: No valid match for {prop} at index {missing_idx}")

    return df
```

```python
import pandas as pd
import numpy as np
import json
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

# Load datasets
train_final = pd.read_csv("train_final.csv")  # Contains features + target
test_final = pd.read_csv("test_final.csv")  # Contains only features
```

```python

# Define target columns
targets = ["HLM_log", "MLM_log", "LogD", "KSOL_scaled", "MDR1-MDCKII_scaled"]

# Define feature set (all columns except targets)
X_train = train_final.drop(['HLM_log', 'HLM', 'MLM_log', 'MLM', 'LogD', 'KSOL_scaled',
                            'KSOL', 'MDR1-MDCKII_scaled', 'MDR1-MDCKII', 'MDR1-MDCKII_log',
                            'fingerprint', 'MW', 'NumHDonors', 'NumHAcceptors', 'LogP'], axis=1)
X_test = test_final.drop(['MW', 'NumHDonors', 'NumHAcceptors', 'LogP'], axis=1)  # Since test_final only contains features
```

```python
X_train.head()
```

```python

# Define hyperparameter grid
param_grid = {
    "n_estimators": [50, 100, 200],
    "max_depth": [None, 10, 20],
    "min_samples_split": [2, 5, 10],
    "min_samples_leaf": [1, 2, 4],
}

# Dictionary to store predictions
predictions_dict = {}

# Loop through each target variable
for column in targets:
    y_train_target = train_final[column]

    # Initialize model
    model = RandomForestRegressor(random_state=42)

    # Perform Grid Search
    grid_search = GridSearchCV(model, param_grid, cv=5, scoring="neg_mean_squared_error", n_jobs=-1)
    grid_search.fit(X_train, y_train_target)

    # Get the best model
    best_model = grid_search.best_estimator_

    # Train on the full dataset with best parameters
    best_model.fit(X_train, y_train_target)

    # Predict on test dataset
    y_pred = best_model.predict(X_test)

    # Store predictions in dictionary
    predictions_dict[column] = y_pred.tolist()

    print(f"{column} - Best Params: {grid_search.best_params_}")

# Save predictions as JSON
with open("predictions.json", "w") as f:
    json.dump(predictions_dict, f)

print("Predictions saved in predictions.json")
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

# Define hyperparameter grid
param_grid = {
    "n_estimators": [50, 100, 200],  # Number of trees
    "max_depth": [None, 10, 20],  # Tree depth
    "min_samples_split": [2, 5, 10],  # Minimum samples to split a node
    "min_samples_leaf": [1, 2, 4],  # Minimum samples per leaf
}

# Lists to store evaluation metrics
mean_squared_errors = []
mean_absolute_errors = []
r2_scores = []

# Target variables
targets = ["HLM_log", "MLM_log", "LogD", "KSOL_scaled", "MDR1-MDCKII_scaled"]

for column in targets:
    y_train_target = y_train[column]
    y_test_target = y_test[column]

    # Initialize model
    model = RandomForestRegressor(random_state=42)

    # Perform Grid Search
    grid_search = GridSearchCV(model, param_grid, cv=5, scoring="neg_mean_squared_error", n_jobs=-1)
    grid_search.fit(X_train, y_train_target)

    # Get best model from GridSearch
    best_model = grid_search.best_estimator_

    # Make predictions on test set
    y_pred = best_model.predict(X_test)

    # Evaluate model performance
    mse = mean_squared_error(y_test_target, y_pred)
    mae = mean_absolute_error(y_test_target, y_pred)
    r2 = r2_score(y_test_target, y_pred)

    mean_squared_errors.append(mse)
    mean_absolute_errors.append(mae)
    r2_scores.append(r2)

    print(f"{column} - Best Params: {grid_search.best_params_}")
    print(f"{column} - Mean Squared Error: {mse:.4f}, Mean Absolute Error: {mae:.4f}, R² Score: {r2:.4f}\n")
```
